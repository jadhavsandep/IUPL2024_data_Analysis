```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Load the deliveries data
deliveries_data = pd.read_csv('deliveries.csv')
```

```python
# Load the matches data
matches_data = pd.read_csv('matches.csv')
```

```python
deliveries_data.head()
```

```python
matches_data.head()
```

```python
# Combine deliveries and matches datasets
combined_data = deliveries_data.merge(matches_data, left_on='match_id', right_on='id', how='inner')
```

```python
combined_data.head()
```

```python
combined_data.shape
```

```python
combined_data['city'].value_counts()
```

```python
combined_data['winner'].value_counts()
```

```python
combined_data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
combined_data[combined_data['is_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)
```

```python
# Calculate total runs scored by each batter
top_run_scorers = combined_data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot the top run-scorers
plt.figure(figsize=(10, 6))
top_run_scorers.plot(kind='bar')
plt.title('Top 10 Run-Scorers Across Matches')
plt.xlabel('Batter')
plt.ylabel('Total Runs Scored')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the number of wickets taken by each bowler
top_wicket_takers = combined_data[combined_data['is_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)

# Plot the top wicket-takers
plt.figure(figsize=(10, 6))
top_wicket_takers.plot(kind='bar')
plt.title('Top 10 Wicket-Takers Across Matches')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets Taken')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total runs and balls faced by each batter
batting_stats = combined_data.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    total_balls=('ball', 'count')
)

# Calculate the strike rate
batting_stats['strike_rate'] = (batting_stats['total_runs'] / batting_stats['total_balls']) * 100

# Filter players with at least 50 balls faced for meaningful analysis
batting_stats_filtered = batting_stats[batting_stats['total_balls'] >= 50].sort_values(by='strike_rate', ascending=False).head(10)

# Plot the top 10 players by strike rate
plt.figure(figsize=(10, 6))
batting_stats_filtered['strike_rate'].plot(kind='bar')
plt.title('Top 10 Batting Strike Rates (Min 50 Balls Faced)')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a column to identify unique partnerships
combined_data['partnership'] = combined_data.apply(
    lambda x: tuple(sorted([x['batter'], x['non_striker']])), axis=1
)

# Calculate total runs scored by each partnership
partnership_stats = combined_data.groupby('partnership')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot the top 10 partnerships by runs scored
plt.figure(figsize=(12, 6))
partnership_stats.plot(kind='bar')
plt.title('Top 10 Partnerships by Runs Scored')
plt.xlabel('Partnership (Player A, Player B)')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze a player's performance trends over matches
# Example: Choose a player (replace 'Player Name' with an actual player name)
player_name = 'BB McCullum'

# Filter the data for the selected player
player_data = combined_data[combined_data['batter'] == player_name]

# Group by match and calculate total runs scored by the player in each match
player_performance = player_data.groupby('match_id')['batsman_runs'].sum()

# Plot the player's performance trend
plt.figure(figsize=(12, 6))
player_performance.plot(kind='line', marker='o')
plt.title(f'Performance Trend of {player_name} (Runs per Match)')
plt.xlabel('Match ID')
plt.ylabel('Total Runs')
plt.grid()
plt.show()
```

```python
# Calculate the total wins by each team across seasons
team_performance = matches_data.groupby(['season', 'winner']).size().unstack(fill_value=0)

# Plot the team performance across seasons
team_performance.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Team Performance Across Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Define the two teams for comparison
team1 = "Kolkata Knight Riders"
team2 = "Mumbai Indians"

# Filter matches won by the two teams
team_comparison = matches_data[matches_data['winner'].isin([team1, team2])]

# Group by season and team to count wins
team_comparison_data = team_comparison.groupby(['season', 'winner']).size().unstack(fill_value=0)

# Plot the comparison
team_comparison_data.plot(kind='bar', figsize=(12, 6))
plt.title(f'Performance Comparison: {team1} vs {team2}')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Teams')
plt.tight_layout()
plt.show()
```

```python
# Define a specific city (replace 'Mumbai' with the desired city)
specific_city = 'Mumbai'

# Filter matches played in the specific city
city_matches = matches_data[matches_data['city'] == specific_city]

# Count wins by teams in the specific city
city_team_performance = city_matches['winner'].value_counts()

# Plot team performances in the specific city
plt.figure(figsize=(12, 6))
city_team_performance.plot(kind='bar', color='skyblue')
plt.title(f'Team Performances in {specific_city}')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Define a specific venue (replace 'Eden Gardens' with the desired venue)
specific_venue = 'Eden Gardens'

# Filter matches played in the specific venue
venue_matches = matches_data[matches_data['venue'] == specific_venue]

# Count wins by teams in the specific venue
venue_team_performance = venue_matches['winner'].value_counts()

# Plot team performances in the specific venue
plt.figure(figsize=(12, 6))
venue_team_performance.plot(kind='bar', color='green')
plt.title(f'Team Wins at {specific_venue}')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter deliveries data for matches where the team won
winning_games = combined_data[combined_data['batting_team'] == combined_data['winner']]

# Calculate total runs contributed by players in winning games
winning_player_contributions = winning_games.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot the top 10 player contributions in winning games
plt.figure(figsize=(12, 6))
winning_player_contributions.plot(kind='bar', color='blue')
plt.title('Top 10 Player Contributions in Winning Games (Total Runs)')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter deliveries data for matches where the bowling team won
winning_bowling_games = combined_data[combined_data['bowling_team'] == combined_data['winner']]

# Calculate total wickets taken by bowlers in winning games
winning_bowler_contributions = winning_bowling_games[winning_bowling_games['is_wicket'] == 1] \
    .groupby('bowler').size().sort_values(ascending=False).head(10)

# Plot the top 10 bowling contributions in winning games
plt.figure(figsize=(12, 6))
winning_bowler_contributions.plot(kind='bar', color='red')
plt.title('Top 10 Bowling Contributions in Winning Games (Total Wickets)')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Analyze match outcomes based on toss decisions
toss_decision_outcomes = matches_data.groupby('toss_decision')['winner'].value_counts().unstack(fill_value=0)

# Plot match outcomes based on toss decisions
toss_decision_outcomes.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Match Outcomes by Toss Decisions')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Wins')
plt.legend(title='Winning Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate total matches played and matches won in each season
matches_per_season = matches_data.groupby('season').size()
wins_per_season = matches_data.groupby('season')['winner'].value_counts().unstack(fill_value=0)

# Calculate win rates for each team by season
win_rates_by_season = (wins_per_season.div(matches_per_season, axis=0) * 100).fillna(0)

# Plot win rates by season
win_rates_by_season.plot(kind='line', marker='o', figsize=(14, 8))
plt.title('Team Win Rates by Seasons')
plt.xlabel('Season')
plt.ylabel('Win Rate (%)')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter data for final matches
final_matches = matches_data[matches_data['match_type'] == 'Final']

# Count the number of finals won by each team
finals_performance = final_matches['winner'].value_counts()

# Plot the team performances in finals
plt.figure(figsize=(12, 6))
finals_performance.plot(kind='bar', color='gold')
plt.title('Team Performances in Finals')
plt.xlabel('Teams')
plt.ylabel('Number of Finals Won')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group final matches by season and count the winners
finals_trends = final_matches.groupby(['season', 'winner']).size().unstack(fill_value=0)

# Plot the trends of finals performances
plt.figure(figsize=(14, 8))
finals_trends.plot(kind='line', marker='o', figsize=(14, 8))
plt.title('Trends of Finals Performances by Teams')
plt.xlabel('Season')
plt.ylabel('Number of Wins in Finals')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total wins by each team in each season
dominant_teams_by_season = matches_data.groupby(['season', 'winner']).size().unstack(fill_value=0)

# Identify the most dominant team in each season
most_dominant_teams = dominant_teams_by_season.idxmax(axis=1)
most_dominant_wins = dominant_teams_by_season.max(axis=1)

# Combine data for visualization
dominant_teams = pd.DataFrame({
    'Most Dominant Team': most_dominant_teams,
    'Wins': most_dominant_wins
})

# Plot the most dominant teams by season
plt.figure(figsize=(12, 6))
dominant_teams['Wins'].plot(kind='bar', color='purple')
plt.title('Most Dominant Teams by Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.xticks(range(len(dominant_teams)), dominant_teams.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Count the number of Player of the Match awards for each player
player_of_match_trends = matches_data['player_of_match'].value_counts().head(10)

# Plot the trends in Player of the Match awards
plt.figure(figsize=(12, 6))
player_of_match_trends.plot(kind='bar', color='orange')
plt.title('Top 10 Players by Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter for maiden overs (total runs in the over is 0)
maiden_overs = combined_data.groupby(['bowler', 'match_id', 'over']).agg(total_runs=('total_runs', 'sum')).reset_index()
maiden_overs = maiden_overs[maiden_overs['total_runs'] == 0]

# Count maiden overs bowled by each bowler
maiden_over_counts = maiden_overs['bowler'].value_counts().head(10)

# Plot the bowlers with the most maiden overs
plt.figure(figsize=(12, 6))
maiden_over_counts.plot(kind='bar', color='blue')
plt.title('Top 10 Bowlers with Most Maiden Overs')
plt.xlabel('Bowler')
plt.ylabel('Number of Maiden Overs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter for deliveries where no runs were scored (dot balls)
dot_balls = combined_data[combined_data['total_runs'] == 0]

# Count the number of dot balls bowled by each player
dot_ball_counts = dot_balls['bowler'].value_counts().head(10)

# Plot the bowlers with the most dot balls
plt.figure(figsize=(12, 6))
dot_ball_counts.plot(kind='bar', color='green')
plt.title('Top 10 Bowlers with Most Dot Balls')
plt.xlabel('Bowler')
plt.ylabel('Number of Dot Balls')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate total runs scored by batters
batting_contributions = combined_data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Calculate total wickets taken by bowlers
bowling_contributions = combined_data[combined_data['is_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)

# Plot batting and bowling contributions
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Plot batting contributions
batting_contributions.plot(kind='bar', color='blue', ax=axes[0])
axes[0].set_title('Top 10 Batting Contributions (Total Runs)')
axes[0].set_xlabel('Player')
axes[0].set_ylabel('Total Runs')
axes[0].tick_params(axis='x', rotation=45)

# Plot bowling contributions
bowling_contributions.plot(kind='bar', color='red', ax=axes[1])
axes[1].set_title('Top 10 Bowling Contributions (Total Wickets)')
axes[1].set_xlabel('Player')
axes[1].set_ylabel('Total Wickets')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Filter for deliveries in the powerplay overs (overs 1-6)
powerplay_overs = combined_data[combined_data['over'].between(1, 6)]

# Calculate total wickets taken by bowlers in powerplay overs
powerplay_bowlers = powerplay_overs[powerplay_overs['is_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)

# Plot the best powerplay bowlers
plt.figure(figsize=(12, 6))
powerplay_bowlers.plot(kind='bar', color='purple')
plt.title('Top 10 Powerplay Bowlers (Total Wickets)')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter for deliveries in the death overs (overs 16-20)
death_overs = combined_data[combined_data['over'].between(16, 20)]

# Calculate total wickets taken by bowlers in death overs
death_over_bowlers = death_overs[death_overs['is_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)

# Plot the best death-over bowlers
plt.figure(figsize=(12, 6))
death_over_bowlers.plot(kind='bar', color='red')
plt.title('Top 10 Death-Over Bowlers (Total Wickets)')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter for deliveries in the death overs (overs 16-20)
death_over_batting = combined_data[combined_data['over'].between(16, 20)]

# Calculate total runs scored by batters in death overs
death_over_batting_performance = death_over_batting.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot the best batting performances in death overs
plt.figure(figsize=(12, 6))
death_over_batting_performance.plot(kind='bar', color='orange')
plt.title('Top 10 Batting Performances in Death Overs (Total Runs)')
plt.xlabel('Batter')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter for deliveries in the powerplay overs (overs 1-6)
powerplay_batting = combined_data[combined_data['over'].between(1, 6)]

# Calculate total runs and balls faced by each batter in powerplay overs
powerplay_stats = powerplay_batting.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    total_balls=('ball', 'count')
)

# Calculate strike rate for powerplay overs
powerplay_stats['strike_rate'] = (powerplay_stats['total_runs'] / powerplay_stats['total_balls']) * 100

# Filter players with at least 20 balls faced for meaningful analysis
powerplay_stats_filtered = powerplay_stats[powerplay_stats['total_balls'] >= 20].sort_values(by='strike_rate', ascending=False).head(10)

# Plot the top 10 strike rates in powerplay overs
plt.figure(figsize=(12, 6))
powerplay_stats_filtered['strike_rate'].plot(kind='bar', color='cyan')
plt.title('Top 10 Strike Rates in Powerplay Overs (Min 20 Balls Faced)')
plt.xlabel('Batter')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
